## **Clustering**

### Optimal K Determination (The Elbow Method)

#### Identify the most appropriate number of clusters (K)

```python
# apply k-means with varying number of clusters from 1 to 6
n = 7
num_clusters = list(range(1, n))

# initialize the Within Cluster Sum of Squares for each run
wcss = []

for k in num_clusters:
    kmeans_model = cluster.KMeans(n_clusters=k, random_state=1)
    kmeans_model.fit(ds)
    wcss.append(kmeans_model.inertia_)

# plot of WCSS versus number of clusters
fig, ax = plt.subplots()
ax.plot(num_clusters, wcss)

ax.set_xlabel('Number of Clusters (k)')
ax.set_ylabel('WCSS')

plt.title('The Elbow for KMeans Clustering')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.9)

plt.show()
```

<div style="background-color: #e6ffe6; border-left: 5px solid #4CAF50; padding: 10px; margin: 10px 0; border-radius: 4px;">
    <strong>Observations:</strong> The Elbow Method graph clearly shows a significant drop in <strong>inertia</strong> between 2 and 4 clusters. This observation led us to choose <strong>K=3</strong> as the optimal number of clusters for this dataset.
</div>

---

### Applying K-Means Clustering Algorithm

**Applying K-Means:** We run the K-means algorithm with the chosen K on our features.

```python
# Initiating the k-means clustering model
kmeans_model = cluster.KMeans(n_clusters = 3, random_state=1)
```

```python
# Fitting k-means using selected features
kmeans_model.fit(ds)
```

#### **Labeling direct features**

```python
# assign the cluster label for each observation
cluster_labels = kmeans_model.labels_

ds = ds.assign(Cluster_ID = cluster_labels)
```

```python
ds.head()
```

---

### Cluster Evaluation

**Average Silhouette Score:**
* Provides an overall measure of the quality of the clustering.
* A higher average silhouette score indicates better-defined and well-separated clusters.
* Values above 0.5 are generally considered good clustering, while values below 0.2 suggest poor clustering.
* The silhouette ranges from −1 to +1, where a positive silhouette score indicates that the object is well-clustered, while a negative score suggests that it might have been assigned to the wrong cluster, as it is closer to a neighboring cluster than to the cluster it is assigned to.

```python
sil_score = metrics.silhouette_score(ds, cluster_labels)
print(f"Average Silhouette Score: {sil_score}")
```

<div style="background-color: #e6ffe6; border-left: 5px solid #4CAF50; padding: 10px; margin: 10px 0; border-radius: 4px;">
    <strong>Quick Comment:</strong> The Average Silhouette Score of 0.638 indicates reasonably well-defined clusters, suggesting good separation and cohesion. While it’s a solid result, there’s still some room for improvement in clustering quality.
</div>

## **3.2 Clustering**

### 3.2.1 Optimal K Determination (The Elbow Method)

#### Identify the most appropriate number of clusters (K)

```python
# apply k-means with varying number of clusters from 1 to 6
n = 7
num_clusters = list(range(1, n))

# initialize the Within Cluster Sum of Squares for each run
wcss = []

for k in num_clusters:
    kmeans_model = cluster.KMeans(n_clusters=k, random_state=1)
    kmeans_model.fit(ds)
    wcss.append(kmeans_model.inertia_)

# plot of WCSS versus number of clusters
fig, ax = plt.subplots()
ax.plot(num_clusters, wcss)

ax.set_xlabel('Number of Clusters (k)')
ax.set_ylabel('WCSS')

plt.title('The Elbow for KMeans Clustering')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.9)

plt.show()
```

<div style="background-color: #e6ffe6; border-left: 5px solid #4CAF50; padding: 10px; margin: 10px 0; border-radius: 4px;">
    <strong>Observations:</strong> The Elbow Method graph clearly shows a significant drop in <strong>inertia</strong> between 2 and 4 clusters. This observation led us to choose <strong>K=3</strong> as the optimal number of clusters for this dataset.
</div>

---

### 3.2.2 Applying K-Means Clustering Algorithm

**Applying K-Means:** We run the K-means algorithm with the chosen K on our features.

```python
# Initiating the k-means clustering model
kmeans_model = cluster.KMeans(n_clusters = 3, random_state=1)
```

### 3.2.3 Cluster Evaluation

**Average Silhouette Score:**
* Provides an overall measure of the quality of the clustering.
* A higher average silhouette score indicates better-defined and well-separated clusters.
* Values above 0.5 are generally considered good clustering, while values below 0.2 suggest poor clustering.
* The silhouette ranges from −1 to +1, where a positive silhouette score indicates that the object is well-clustered, while a negative score suggests that it might have been assigned to the wrong cluster, as it is closer to a neighboring cluster than to the cluster it is assigned to.

```python
sil_score = metrics.silhouette_score(ds, cluster_labels)
print(f"Average Silhouette Score: {sil_score}")
```
